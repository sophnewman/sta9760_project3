```python
# import libraries
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stock_data = pd.read_csv('results.csv')
stock_data.head()
```

```python

plot1 = sns.catplot(x='hour',y='hourly_high', hue = 'company',
           data = stock_data,
           kind = 'bar',
                    edgecolor = 'black',
                   height = 9)


plot1.set(xlabel = "Hour", ylabel = 'Hourly High (USD)')
plot1.set(title = 'Hourly Highs per Stock (USD)')
```

```python
# line plots of each company, side by side

sns.lineplot(x='hour',y='hourly_high', hue = 'company',
           data = stock_data)

plot1.set(xlabel = "Hour", ylabel = 'Hourly High (USD)')
plot1.set(title = 'Hourly Highs per Stock (USD)')
plt.legend(bbox_to_anchor=(1.4,1), loc = 'upper right')
```

```python
# individual line plots for each ticker
# shows a trend line for pricing thoughout the day

stock_list = stock_data['company'].unique()
for stock in stock_list:
    image = stock_data[stock_data['company'] == stock]
    plt.plot(image.hour, image.hourly_high, color = 'red')
    plt.title(f"{company} -- Hourly Price High")
    plt.xlabel("Hour")
    plt.ylabel("Hourly High Price")
    plt.show()
```
